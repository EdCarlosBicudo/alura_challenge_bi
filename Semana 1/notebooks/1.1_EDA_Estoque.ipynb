```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Análise Exploratória de Dados (EDA) da tabelas.

### Faço uma análise de cada tabela para entender melhor como são constituídas e se algum campo precisa de tratamento.

## Primeiro analiso a tabela de estoque.

```python
estoque = pd.read_csv('../dados/Tabela - Estoque.csv')
```

```python
estoque.head()
```

### Aqui vemos que não há valores faltantes na tabela

```python
estoque.info()
```

### Com um resumo do campo 'Quantidade' vemos como ela se comporta, e vemos que nenhum produto chegou a ficar sem estoque.

```python
estoque['Quantidade'].describe()
```

### Aqui vemos graficamente o comportamento do campo 'Quantidade'

```python
ax = sns.boxplot(x = 'Quantidade', data = estoque, orient = 'h')
ax.figure.set_size_inches(18, 4)
ax.set_title("Estoque dos Produtos", fontsize = 18)
ax.set_xlabel("Quantidade", fontsize = 14)
ax.set_ylabel("Produtos", fontsize = 14)
plt.show()
```

### Aqui vemos o comportamento da 'Quantidade' por produto.

Mesmo sem saber o que é cada produto, vemos que o estoque de cada produto se comporta de maneira muito diferente.

```python
ax = sns.boxplot(x = 'Quantidade', y= "ID Produto", data = estoque, orient = 'h')
ax.figure.set_size_inches(20, 20)
ax.set_title("Estoque dos Produtos", fontsize = 18)
ax.set_xlabel("Quantidade", fontsize = 14)
ax.set_ylabel("ID Produto", fontsize = 14)
plt.show()
```
