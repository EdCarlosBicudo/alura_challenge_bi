```python
import pandas as pd
import numpy as np
import locale
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
```

# Análise dos Valores

## Devido aos valore das dua Tabelas não parecerem consistentes estou fazendo essa análise para entender como se relacionam.

```python
dados_dispositivos = pd.read_json('../dados/Tabela_dispositivos.json')
dados_idade_genero = pd.read_json("../dados/Tabela_idade_e_genero.json")
```

### Como os dados da Tabela estão no padrão pt_BR, com a ',' como separador decimal, o python entende os valor como String, e não como Float.

Por isso faço o processo de localização para serem entendidos como Float pelo python

```python
def convert(value):
    if not value:
        return np.nan
    return locale.atof(value)
```

```python
dados_dispositivos['Quantia gasta (BRL)'] = dados_dispositivos['Quantia gasta (BRL)'].map(convert)
dados_dispositivos['Cliques em links'] = dados_dispositivos['Cliques em links'].map(convert)
dados_dispositivos['Visualizações por página'] = dados_dispositivos['Visualizações por página'].map(convert)
dados_dispositivos['Compras no website'] = dados_dispositivos['Compras no website'].map(convert)
dados_dispositivos['Compras no facebook'] = dados_dispositivos['Compras no facebook'].map(convert)
dados_dispositivos['Adicionados ao carrinho'] = dados_dispositivos['Adicionados ao carrinho'].map(convert)
dados_dispositivos['Valor de conversão adicionado ao carrinho'] = dados_dispositivos['Valor de conversão adicionado ao carrinho'].map(convert)
dados_dispositivos['Checkouts Iniciados'] = dados_dispositivos['Checkouts Iniciados'].map(convert)
dados_dispositivos['Valor de conversão de checkouts iniciados'] = dados_dispositivos['Valor de conversão de checkouts iniciados'].map(convert)
dados_dispositivos['Valor de conversão de compras'] = dados_dispositivos['Valor de conversão de compras'].map(convert)
dados_dispositivos['Compras'] = dados_dispositivos['Compras'].map(convert)
dados_idade_genero['Quantia gasta (BRL)'] = dados_idade_genero['Quantia gasta (BRL)'].map(convert)
dados_idade_genero['Cliques no link'] = dados_idade_genero['Cliques no link'].map(convert)
dados_idade_genero['Visualizações por página'] = dados_idade_genero['Visualizações por página'].map(convert)
dados_idade_genero['Compras no website'] = dados_idade_genero['Compras no website'].map(convert)
dados_idade_genero['Compras no facebook'] = dados_idade_genero['Compras no facebook'].map(convert)
dados_idade_genero['Adicionados ao carrinho'] = dados_idade_genero['Adicionados ao carrinho'].map(convert)
dados_idade_genero['Valor de conversão adicionado ao carrinho'] = dados_idade_genero['Valor de conversão adicionado ao carrinho'].map(convert)
dados_idade_genero['Checkouts Iniciados'] = dados_idade_genero['Checkouts Iniciados'].map(convert)
dados_idade_genero['Valor de conversão de checkouts iniciados'] = dados_idade_genero['Valor de conversão de checkouts iniciados'].map(convert)
dados_idade_genero['Valor de conversão de compras'] = dados_idade_genero['Valor de conversão de compras'].map(convert)
dados_idade_genero['Compras'] = dados_idade_genero['Compras'].map(convert)
```

## Depois de explorar os dados de um único dia consegui entender como os dados da Tabela 'Dispositívos' são agregados.

Os números de Compras parecem duplicados, filtrando as duplicidades chegamos aos mesmos valores.

```python
dados_dispositivos[dados_dispositivos['Dia']=='2021-07-31'][['Plataforma do dispositivo','Colocação','Plataforma','Compras']]
```

```python
soma_compras_dia_por_dispositivo = dados_dispositivos[(dados_dispositivos['Dia']=='2021-07-31')&
                                                      (dados_dispositivos['Plataforma do dispositivo']=='All')&
                                                      (dados_dispositivos['Colocação']=='All')&
                                                      (dados_dispositivos['Plataforma']=='All')]['Compras'].sum()
soma_compras_total_dispositivos = dados_dispositivos[(dados_dispositivos['Plataforma do dispositivo']=='All')&
                                                     (dados_dispositivos['Colocação']=='All')&
                                                     (dados_dispositivos['Plataforma']=='All')]['Compras'].sum()
soma_compras_dia_idade_genero = dados_idade_genero[dados_idade_genero['Dia']=='2021-07-31']['Compras'].sum()
soma_compras_total_idade_genero = dados_idade_genero['Compras'].sum()
```

```python
print(f"Total de Vendas pela Tabela Dispositivos, no dia 31-7-2021: {soma_compras_dia_por_dispositivo}")
print(f"Total de Vendas pela Tabela Idade e Gênero, no dia 31-7-2021: {soma_compras_dia_idade_genero}")
print(f"Total de Vendas da Tabela Dispositivos: {soma_compras_total_dispositivos}")
print(f"Total de Vendas da Tabela Idade e Gênero: {soma_compras_total_idade_genero}")
```
